```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

### 1. Which percentage of airline passengers are satisfied? Does it vary by customer type? What about type of travel?

```python
round(df["Satisfaction"].value_counts(normalize = True)*100)
```

#### About 43% passengers are sataisfied

```python
sns.catplot(x = "Customer Type", data = df,kind = "count", hue = "Satisfaction")
plt.xticks(rotation = 45)
```

```python
round(df.groupby("Customer Type")["Satisfaction"].value_counts(normalize = True)*100)
```

```python
df["Customer Type"].value_counts()
```

```python
sns.catplot(x = "Satisfaction", data = df,kind = "count", hue = "Customer Type")
plt.xticks(rotation = 45)
```

#### Yes Customer satisfaction vary by Customer type
 Only 24% of First Time Customers are satisfied and 48% of Returning Customers are satisfied.

```python
round(df.groupby("Type of Travel")["Satisfaction"].value_counts(normalize = True)*100)
```

```python
col_palette= {"Satisfied": "Green", "Neutral or Dissatisfied": "Red"}
sns.catplot(x = "Type of Travel", data = df,kind = "count",col = "Customer Type", hue = "Satisfaction", palette = col_palette, alpha = 0.7)
plt.xticks(rotation = 45)
```

```python
col_palette= {"Satisfied": "Green", "Neutral or Dissatisfied": "Red"}
sns.catplot(x = "Customer Type", data = df,kind = "count",col = "Type of Travel", hue = "Satisfaction", palette = col_palette, alpha = 0.5)
plt.xticks(rotation = 45)
```
